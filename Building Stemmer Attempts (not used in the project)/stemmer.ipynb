```python
#preprocessing function
def PreprocessingFunct(liste):
  preprocessed = []
  for i in range(len(liste)):
    lowercase = str(liste[i]).lower()
    nopunct = lowercase.strip(".").strip(",").strip("!").strip("?").strip("'").strip(":").strip(";")
    preprocessed.append(nopunct.split())
  return preprocessed
```
