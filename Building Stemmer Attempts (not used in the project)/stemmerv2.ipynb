```python
#create a dataframe
import pandas as pd
df = pd.read_excel ("kadin_surucu.xlsx") 
print (df)
```

```python
#extract the texts as a list
metinler = df["Metinler"].tolist()
print(metinler)
```

```python
#we need this for stemming
!pip install TurkishStemmer
from TurkishStemmer import TurkishStemmer
```

```python
stemmer = TurkishStemmer()

sentstemmed = []
for i in range(len(metinler)):
  sentence = str(metinler[i])
  sentence = sentence.split()
  for word in sentence:
    #it is not a great stemmer so it analyses sürücü as sürüç.ACC. that is why i introduced
    #a list of exceptions. i will sort starlang's much better morph disambiguator out soon
    #so we wont be needing this. or idk i will create a new stemmer, we will see.
    edges1 = ["sürücü", "sürücüyü", "sürücüye", "sürücüde", "sürücüden", "sürücüyle", "sürücüler", "sürücünün"]
    edges2 = ["kadın", "kadını", "kadına", "kadında", "kadından", "kadınla", "kadınlar", "kadının"]
    if word in edges1:
      sentstemmed.append("sürücü")
    if word in edges2:
      sentstemmed.append("kadın")
    else:   
      word = stemmer.stem(word)
      sentstemmed.append(word)

print(sentstemmed)


```
